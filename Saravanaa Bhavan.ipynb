# Saravanaa Bhavan Branches

## About
#### Saravanaa Bhavan is the largest South Indian restaurant chain in the world. It's got branches all over the world. Just out of curiosity I wanted to visualize where they were located on a map. Since I couldn't find one, I decided to create one for myself.

#### I used Selenium to get the content from their website and create a dataset. And subsequently plotted the data on the map.

```python
import pandas as pd # read data
import folium # mplot points on map
```

```python
branchesDF = pd.read_csv("E:/My Jupyter/Data Sets/SaravanaaBhavanBranches.csv")
branchesDF.head(5)
```

```python
centreLat = branchesDF["Latitude"].mean()
centreLon = branchesDF["Longitude"].mean()
# create a map
globalMap = folium.Map(location = [centreLat,centreLon], 
                    zoom_start = 2,
                    scrollWheelZoom = False)
globalMap # display map
```

```python
# plotting points on the map
for i in branchesDF.index:
    branchLat = branchesDF["Latitude"][i]
    branchLon = branchesDF["Longitude"][i]
    iconColor = "green"
    address = branchesDF["Address"][i]
    if(branchesDF["Country"][i] == "TO BE OPENED"):
        iconColor = "orange"
        address = address + " : To Be Opened"
    folium.Marker(
        location = [branchLat, branchLon],
        tooltip = address,
        icon = folium.Icon(icon = "", color = iconColor)
    ).add_to(globalMap)
```
